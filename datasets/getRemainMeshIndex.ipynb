# Get the index of the mesh that are useful
- The head mesh (exclude neck)
- Remove eye ball

```python
import numpy as np
import plotly.graph_objs as go
import os
```

```python
root_dir = "./kuoyuan_home_glass"
deca_deform_example_path = os.path.join(root_dir, "deca_deform_out/p1_000000/p1_000000_verts.txt")
remove_index_path = "./use_mesh_indexes.npy"
```

```python
deca_deformed_mesh = [] 
with open(deca_deform_example_path,"r") as mesh_file:
    lines = mesh_file.readlines()
    for line in lines:
        points = line.strip().split(" ")
        deca_deformed_mesh.append(points)
deca_deformed_mesh = np.array(deca_deformed_mesh,dtype=np.float32)
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter3d(
    x=deca_deformed_mesh[:, 0],
    y=deca_deformed_mesh[:, 1],
    z=deca_deformed_mesh[:, 2],
    mode='markers',
    marker=dict(size=2),
))
fig.update_layout(scene_dragmode='orbit')
fig.show()
```

```python
to_remove_indices_1 = np.where((deca_deformed_mesh[:, 1] > 0.31))[0]
```

```python
to_remove_indices_2 = np.where((deca_deformed_mesh[:, 1] > 0.2755) & (deca_deformed_mesh[:, 2] > 29.38))[0]
```

```python
to_remove_indices_3 = np.where((deca_deformed_mesh[:, 0] > -0.09388) & (deca_deformed_mesh[:, 0] < -0.02825) & (deca_deformed_mesh[:, 1] > -0.07256) & (deca_deformed_mesh[:, 1] <-0.005718) &(deca_deformed_mesh[:, 2] > 29.14086) & (deca_deformed_mesh[:, 2] < 29.1936))[0]
```

```python
to_remove_indices_4 = np.where((deca_deformed_mesh[:, 0] > 0.0735) & (deca_deformed_mesh[:, 0] < 0.1419835) & (deca_deformed_mesh[:, 1] > -0.0623) & (deca_deformed_mesh[:, 1] <0.00448) &(deca_deformed_mesh[:, 2] > 29.12675) & (deca_deformed_mesh[:, 2] < 29.18045))[0]
```

```python
all_indices = np.arange(0,len(deca_deformed_mesh))
```

```python
remain_indices = np.setdiff1d(all_indices, to_remove_indices_1)
remain_indices = np.setdiff1d(remain_indices, to_remove_indices_2)
remain_indices = np.setdiff1d(remain_indices, to_remove_indices_3)
remain_indices = np.setdiff1d(remain_indices, to_remove_indices_4)
```

```python
deca_deformed_mesh_filter = deca_deformed_mesh[remain_indices]
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter3d(
    x=deca_deformed_mesh_filter[:, 0],
    y=deca_deformed_mesh_filter[:, 1],
    z=deca_deformed_mesh_filter[:, 2],
    mode='markers',
    marker=dict(size=2),
))
fig.update_layout(scene_dragmode='orbit')
fig.show()
```

```python
np.save(remove_index_path,remain_indices)
```
